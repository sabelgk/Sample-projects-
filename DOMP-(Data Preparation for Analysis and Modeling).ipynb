# Housing Days On Market - Data preparation for Analysis and Modeling 

## Information

The features gathered from different sources were combined and wrangled using SQLite and python and placed in SQLite database. ‘Raw_Data.csv‘is the output of these compiled and wrangled data, it originally has 51,720 instances. 

This jupyter notebook inspect and cleans the compiled raw dataset (Raw_Data.csv) further to make it ready for modeling and analysis. We have changed the list date and close dates to date time data type; converted features to integer or floats; removed instances with missing features; label encoded categorical features, and limited the instances to be from list date -January 2011 to December 2014 to align all of the features in our dataset. 

‘df_4.csv’ is the output of the inspection and cleaning with 11,361 instances and 117 features.

## Convert list date and close date to date time data type

```python
# Separate out year, month, and day from ListDate and CloseDate
df_2['CloseDate_dt'] = pd.to_datetime(df_2['CloseDate'])
df_2['CloseDate_dt_year'] = df_2['CloseDate_dt'].dt.year
df_2['CloseDate_dt_month'] = df_2['CloseDate_dt'].dt.month
df_2['CloseDate_dt_day'] = df_2['CloseDate_dt'].dt.day
df_2['ListDate_dt'] = pd.to_datetime(df_2['ListDate'])
df_2['ListDate_dt_year'] = df_2['ListDate_dt'].dt.year
df_2['ListDate_dt_month'] = df_2['ListDate_dt'].dt.month
df_2['ListDate_dt_day'] = df_2['ListDate_dt'].dt.day
```

```python
#Converting the closed date and list date from date time to integer so we can be abel to analyze it
df_2['CloseDate_all']=1000*df_2['CloseDate_dt_year']+100*df_2['CloseDate_dt_month']+df_2['CloseDate_dt_day']
df_2['ListDate_all']=1000*df_2['ListDate_dt_year']+100*df_2['ListDate_dt_month']+df_2['ListDate_dt_day']
df_2.head()
```

```python
# Drop original Date columns and the columns we have changed to date since we have a date column.
df_2 = df_2.drop(u'CloseDate', axis = 1)
df_2 = df_2.drop(u'ListDate', axis = 1)
```

## Date used for analysis and modeling (2010-2014)

```python
# Set date range for analysis and modelling
df_2 = df_2[(df_2['ListDate_dt_year'] >= 2010) & (df_2['CloseDate_dt_year'] <= 2014)]
```

```python
from sklearn.base import BaseEstimator, TransformerMixin 
from sklearn.preprocessing import LabelEncoder

CATEGORICAL = ['ES_IsCharter', 'ML#','TaxTotalLivingArea','mimiStatus','BasementY/N','ES_IsMagnet', 'ES_IsTitleI', 'ES_IsTitleI.1', 'ES_IsVirtual', 'HS_IsCharter', 'HS_IsMagnet', 'HS_IsTitleI', 'HS_IsTitleI.1', 'HS_IsVirtual', 'MS_IsCharter', 'MS_IsMagnet', 'MS_IsTitleI', 'MS_IsTitleI.1', 'MS_IsVirtual']
for col in ['ES_IsCharter', 'ML#', 'TaxTotalLivingArea','mimiStatus','BasementY/N','ES_IsMagnet', 'ES_IsTitleI', 'ES_IsTitleI.1', 'ES_IsVirtual', 'HS_IsCharter', 'HS_IsMagnet', 'HS_IsTitleI', 'HS_IsTitleI.1', 'HS_IsVirtual', 'MS_IsCharter', 'MS_IsMagnet', 'MS_IsTitleI', 'MS_IsTitleI.1', 'MS_IsVirtual']:
      df_4[col] = df_4[col].apply(str)

class EncodeCategorical(BaseEstimator, TransformerMixin):
    
    def __init__(self, columns=None):
        self.columns = list(columns) if columns is not None else None
        self.encoders = None 
    
    def fit(self, DFM_2, target=None):
        if self.columns is None:
            self.columns = list(df_4.columns)
        
        self.encoders = {
            column: LabelEncoder().fit(DFM_2[column])
            for column in self.columns
        }
        
        return self
    
    def transform(self, df_4):
        dF_4 = df_4.copy()
        
        for column, encoder in self.encoders.items():
            df_4[column] = encoder.transform(df_4[column])
        
        return df_4

df_4=EncodeCategorical(CATEGORICAL).fit_transform(df_4)
df_4.info()
```
